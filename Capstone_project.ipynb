### Task 1: Create a dataframe about Toronto's neighborhoods <a name="PA_2_1" ></a>
To solve this task, we have to scrape the neighborhoods table from Toronto's [Wikipedia website](https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M). <br><br>

To do so, we will use the Pandas "read_html" function in order to scrape the Wikipedia page.

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M" # Define url
dfs = pd.read_html(url) # Download tables on website via pandas
print("The downloaded website contains {} tables.".format(len(dfs))) # Display how many tables have been downloaded
```

As we can see, we have downloaded multiple tables. A quick visual check on the website itself shows us that the information we are interested in lies within table 1. We thus need to extract table 1 from the downloaded website.

```python
df = dfs[0] # Get first table
df.head()
```

We can see that we downloaded the correct data. The data, however, is messy and needs to be cleaned before we can use it for something else:

#### Stacking the columns horizontally
To be able to iterate through the columns more easily, we need to stack them. We can do so by appending each column to a new variable. In result, we will have all converted our initial df with 20 rows and 9 columns to a new df with 180 rows and 1 column.

```python
df_stacked = df[0] # Grab first column in new variable

for column in df.columns[1:]: # Iterate through columns, starting with the second column
    df_stacked = df_stacked.append(df[column]) # Append each column to new variable

df_stacked = pd.DataFrame(df_stacked) # Convert back to df
print("The stacked df has {} rows and {} columns:".format(df_stacked.shape[0], df_stacked.shape[1]))
df_stacked.head()
```

#### Getting the post codes
When we observe the data more closely, we can notice that the postal code is contained in the first three letters of each string. We can store these in a new column by accessing the first three positions of the string:

```python
df_stacked['POSTAL_CODE'] = df_stacked[0].str[:3] # Receive first three characters as string
df_stacked.head()
```

#### Getting the boroughs
When we have a second look at the downloaded information, we can see that the boroughs always start after the postal code (first three letters) and end before a "(" character. We can use the split function to split our information and then delete the rest of the string to isolate the boroughs:

```python
split_char = "(" # Define ( as the character to split the string
df_stacked['BOROUGH'] = df_stacked[0].str.split(pat=split_char) # Splits the string after the ( character

df_stacked['BOROUGH'] = df_stacked['BOROUGH'].str[0] # Select first split
df_stacked['BOROUGH'] = df_stacked['BOROUGH'].str[3:] # Kick out the first three letters (postal code)

df_stacked.head()
```

#### Getting the neighborhoods
In this step, we need to extract the neighborhoods. We can do that with the same approach as above. Afterwards, we will seperate the neighborhoods with a "," instead of a "/" as the course material tells us to.

```python
split_char = "(" # Define ( as the character to split the string
df_stacked['NEIGHBORHOOD'] = df_stacked[0].str.split(pat=split_char) # Splits the string after the "(" character

df_stacked['NEIGHBORHOOD'] = df_stacked['NEIGHBORHOOD'].str[1] # Select second split
df_stacked['NEIGHBORHOOD'] = df_stacked['NEIGHBORHOOD'].str[:-1] # Kick out the last character "("

df_stacked['NEIGHBORHOOD'] = df_stacked['NEIGHBORHOOD'].str.replace(" / ", ", ")

df_stacked.head()
```

#### Cleaning up the dataframe
Finally, we can clean up the data to fullfill the courses' requirements for task 1. First, we will drop the column with the messy data:

```python
df_stacked.drop(columns=[0], inplace=True)
df_stacked.head()
```

Next, we will get rid of all rows where the borough is not assigned:

```python
df_stacked = df_stacked.reset_index(drop=True) # Reset index
index_names = df_stacked[df_stacked['BOROUGH'] == 'Not assigned'].index # Define condition
df_stacked.drop(index_names, axis=0, inplace=True) # Drop rows that meet condition
df_stacked.head()
```

The course says that the data set includes duplicates in the postal codes column. We therefore will check our dataframe for duplicates: 

```python
duplicates = df_stacked['POSTAL_CODE'].duplicated() # Check for duplicates
print("We have {} duplicates in our postal codes.".format(duplicates.sum())) # Count duplicates = True
```

As we have no duplicates in our postal code column, we can move on and check whether we have a the value "Not assigned" in our neighborhoods column:

```python
if 'Not assigned' in set(df_stacked['NEIGHBORHOOD']) == True:
    print("There are missing neighborhoods, please check.")

else:
    print("No missing neighborhoods, proceed.")
```

Since we do not have any unassigned neighborhoods, we can proceed to the final step - assigning our cleaned dataframe to a new variable for further processing:

```python
toronto = df_stacked
print("The Torono neighborhood df has {} rows and {} columns:".format(toronto.shape[0], toronto.shape[1]))
print()
toronto.head()
```

### Visualizing the Toronto's neighborhoods
Before conducting any kind of analysis, we will visualize Toronto's boroughs via folium. To do so, we will first assign a unique number for each borough so that our map gets a better readbility:

```python
# Create borough ID
toronto['BOROUGH_ID'] = toronto.groupby(['BOROUGH']).ngroup()
print("Toronto has {} unique boroughs:".format(len(toronto['BOROUGH_ID'].unique())))
print()

for borough in sorted(toronto['BOROUGH'].unique()):
    print(borough)
```

Oh no! What happened here?!?! Apparently, we have scraped some false information. Our data set seems to include processing centres for Toronto's post. Let us have a look at this before we continue:

```python
# Function to pretty up printing
def print_full(df):
    pd.set_option('display.max_rows', len(df))
    print(df.to_string(index=False))
    print()
    pd.reset_option('display.max_rows')

# Print the outliers
print_full(toronto[(toronto['BOROUGH'] == 'Downtown TorontoStn A PO Boxes25 The Esplanade')])
print_full(toronto[(toronto['BOROUGH'] == 'East TorontoBusiness reply mail Processing Centre969 Eastern')])
print_full(toronto[(toronto['BOROUGH'] == 'East YorkEast Toronto')])
print_full(toronto[(toronto['BOROUGH'] == 'EtobicokeNorthwest')])
print_full(toronto[(toronto['BOROUGH'] == 'MississaugaCanada Post Gateway Processing Centre')])
```

Alright, "East YorkEast Toronto" and "EtobicokeNorthwest" have neighborhoods stored in the data set. The others, however, are clearly no actual boroughs but processing centers for Toronto's post service.<br>
A quick Google search confirms this theory. However, "East Toronto" is not its own borough but belongs to "East York". Same goes for "Etobicoke". <br>
We will therefore replace these with their actual names and kick out the processing centers:

```python
# Reset index for toronto df
toronto.reset_index(inplace=True)

# Define false entrees
false_entrees = [
    'Downtown TorontoStn A PO Boxes25 The Esplanade',
    'East TorontoBusiness reply mail Processing Centre969 Eastern',
    'MississaugaCanada Post Gateway Processing Centre']

# Loop list and drop respective entrees:
for entry in false_entrees:
    false_index = toronto[(toronto['BOROUGH'] == entry)].index # Get index for rows with false entries
    toronto.drop(false_index, inplace=True) # Drop respective rows

# Replace wrong borough names
toronto.replace('East YorkEast Toronto', 'East York', inplace=True)
toronto.replace('EtobicokeNorthwest', 'Etobicoke', inplace=True)
```

Let's check, if that worked by performing the above step again: 

```python
# Create borough ID
toronto['BOROUGH_ID'] = toronto.groupby(['BOROUGH']).ngroup()
print("Toronto has {} unique boroughs:".format(len(toronto['BOROUGH_ID'].unique())))
print()

for borough in sorted(toronto['BOROUGH'].unique()):
    print(borough)
```

Perfect. Let's move on to the map! For better visualization, each borough ID gets a unique colour. We will also add some html to the popup to make it look better:

```python
# Create colour scheme for borough IDs
x = np.arange(len(toronto['BOROUGH_ID'].unique())) # Define number of groups
ys = [i + x + (i*x)**2 for i in range(len(toronto['BOROUGH_ID'].unique()))] # Create a unique numpy array for each ID
colors_array = cm.gist_rainbow(np.linspace(0, 1, len(ys))) # Convert arrays to float between 0 and 1 in chosen colour scheme
rainbow = [colors.rgb2hex(i) for i in colors_array] # Convert floats to hex code

# Define starting point for map
address = 'Toronto, Ontario' # Enter address
geolocator = Nominatim(user_agent="toronto_analysis") # Initialize with individual agent
location = geolocator.geocode(address) # Get address object
latitude = location.latitude # Get lat
longitude = location.longitude # Get lng

# Create map of Toronto
map_toronto = folium.Map(
    location=[latitude, longitude], # Define starting point
    zoom_start=10, # Select starting zoom
    tiles='OpenStreetMap' # Choose map look
    )  

# Add markers to map
for lat, lng, borough, borough_id, neighborhood in zip(toronto['LATITUDE'], toronto['LONGITUDE'], toronto['BOROUGH'], toronto['BOROUGH_ID'], toronto['NEIGHBORHOOD']): # Pass information
    popup_text = '<u><b>Boroughs:</u></b><br>{}<br><br><u><b>Neighborhoods:</u></b><br>{}'.format(borough, neighborhood) # Define text for popup
    popup_text = folium.Popup(popup_text) # Add text to map
    folium.CircleMarker( # Create circles
        [lat, lng], # Parse position
        radius=5, # Define radius
        popup=popup_text, # Parse pop up text
        color=rainbow[borough_id-1], # Iterate and parse hexcode
        fill=True, # Define fill
        fill_color=rainbow[borough_id-1], # Iterate and parse hexcode
        fill_opacity=0.5, # Define opacity
        parse_html=False # Popup does not need to be rendered
        ).add_to(map_toronto) # Add circles to map 

map_toronto # Display
```

We can see that the boroughs (and thus the neighboorhoods) are becomming more and more dense towards the city centre. As we have two dogs, we need space and parks to walk them. We are threfore excluding everything in the city centre and assign the remaining borough to our target group:

```python
# Store interesting boroghs in new variable
target_boroughs = toronto[
    (toronto['BOROUGH'] != "Downtown Toronto") &
    (toronto['BOROUGH'] != "West Toronto") &
    (toronto['BOROUGH'] != "Queen's Park") &
    (toronto['BOROUGH'] != "East Toronto") &
    (toronto['BOROUGH'] != "East York") &
    (toronto['BOROUGH'] != "Central Toronto") &
    (toronto['BOROUGH'] != "York") 
    ]
```
